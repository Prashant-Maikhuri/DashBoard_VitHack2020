**BLOOD PRESSURE**

```
import pandas as pd
import random
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.layers import Dense,Input,LSTM
from tensorflow.keras.models import Model
%matplotlib inline
```

```
bm_norm=np.arange(80,120,0.01)
bm_norm.shape
bm_high=np.arange(120,140,0.01)
bm_norm.shape
```

```
def train():
  b=[]
  for h in range(0,188):
    k=np.expand_dims(random.choice(bm_high),-1)
    b.append(k)
    
  return b   
```

```
def train_():
  a=[]
  for i in range(0,188):
    k=np.expand_dims(random.choice(bm_norm),-1)
    a.append(k)
    
  return a
```

```
xtrain=np.zeros((600,188,1))
ytrain=np.zeros((600,1))
for j in range(0,200):
   xtrain[j,:,:]=train()
   ytrain[j,:]=np.array(1)
```

```
for i in range(200,600):
   xtrain[j,:,:]=train_()
   ytrain[j,:]=np.array(0)
   
```

```
plt.plot(xtrain[10])
plt.xlabel("time")
plt.ylabel("Blood Pressure")
```

```
i=Input(shape=(xtrain[0].shape))
x=LSTM(128)(i)
x=Dense(1,activation="sigmoid")(x)

model=Model(i,x)
```

```
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```
r=model.fit(xtrain,ytrain,validation_split=0.2,epochs=2)
```
