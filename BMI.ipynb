```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input,Dense,LSTM
from tensorflow.keras.models import Model
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
import pandas as pd
from tensorflow.keras.optimizers import Adam
```

```python
bmi=pd.read_csv(r"E:\BMI\500_Person_Gender_Height_Weight_Index.csv")
print(bmi)
```

```python
colum=["Height","Weight"]
df=pd.read_csv(r"E:\BMI\500_Person_Gender_Height_Weight_Index.csv",usecols=colum)
print(df)
col_index=["Index"]
df2=pd.read_csv(r"E:\BMI\500_Person_Gender_Height_Weight_Index.csv",usecols=col_index)
print(df2)
```

```python
index=tf.one_hot(df2["Index"],6)
index[0:3]
```

```python
df.iloc(axis=0)[0]
```

```python
xtrain=np.array(df.iloc(axis=0)[0],df.iloc(axis=1)[1])
ytrain=np.array(index)
```

```python
names=["Height","Weight"]
xtrain=np.array(df[names])
xtrain.shape
```

```python
scaler=StandardScaler()
h=scaler.fit_transform(xtrain)
g=scaler.fit_transform(ytrain)
```

```python
plt.plot(xtrain)
```

```python
i=Input(shape=xtrain[0].shape)
x=Dense(128,activation="tanh")(i)
x=Dense(6,activation="softmax")(x)
model=Model(i,x)
```

```python
r=model.compile(optimizer=Adam (learning_rate=0.0001),loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
b=model.fit(xtrain,ytrain,validation_split=0.45,epochs=999)
```

```python

plt.plot(b.history['loss'])
plt.plot(b.history['val_loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.title('model loss')
plt.show()
```

```python
plt.plot(b.history['accuracy'])
plt.plot(b.history['val_accuracy'])
plt.title('model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.show()
```
